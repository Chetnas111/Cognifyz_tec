number of votes and the rating of a
restaurant.


```python
#1.Identify the restaurants with the highest and lowest number of votes
df.sort_values(by='Votes', ascending=False)
```

```python
highest_votes=df.groupby('Restaurant Name')['Votes'].sum().reset_index(name='Total_votes')
highest_votes=highest_votes.sort_values(by='Total_votes', ascending=False).iloc[0]
print(highest_votes)
```

```python
lowest_votes=df.groupby('Restaurant Name')['Votes'].sum().reset_index(name='Total_votes')
lowest_votes=lowest_votes.sort_values(by='Total_votes', ascending=True).iloc[0]
print(lowest_votes)
```

```python
#2.Analyze if there is a correlation between the number of votes and the rating of a restaurant.
correlation=np.corrcoef(df['Votes'],df['Aggregate rating'])
correlation=np.round(correlation,2)
print(correlation)
```

#Task3: Price Range vs. Online Delivery and Table Booking
-- 1.Analyze if there is a relationship between the
price range and the availability of online
delivery and table booking.

-- 2.Determine if higher-priced restaurants are
more likely to offer these services

```python
#1.Analyze if there is a relationship between the price range and the availability of online delivery and table booking.
plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
sns.boxplot(x='Has Online delivery', y='Average Cost for two', data=df)
plt.xlabel('Has Online delivery')
plt.ylabel('Average Cost for two')
plt.title('Affect of cost on online delivery \n')

plt.subplot(1,2,2)

sns.boxplot(x='Has Table booking', y='Average Cost for two', data=df )
plt.xlabel('Table booking')
plt.ylabel('Average Cost for two')
plt.title('Affect of Cost on Table Booking \n')

plt.tight_layout()
plt.show()
```

```python
# 2.Determine if higher-priced restaurants are more likely to offer these service
plt.figure(figsize= (10,5))
plt.subplot(1,2,1)
sns.boxplot(x='Has Table booking', y='Average Cost for two', data=df.loc[df['Average Cost for two']<10000 , :] )
plt.title('Affect of cost on Table booking')
plt.subplot(1,2,2)
sns.boxplot(x='Has Online delivery', y='Average Cost for two', data=df.loc[df['Average Cost for two']<10000 , :] )
plt.title('Affect of cost on Online delivery')
plt.tight_layout()
plt.show()
```
